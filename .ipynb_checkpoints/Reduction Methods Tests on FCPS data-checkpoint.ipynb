## Import FCPS data

```python
atom=pd.read_csv("Data/atom.csv")
Chainlink=pd.read_csv("Data/Chainlink.csv")
EngyTime=pd.read_csv("Data/EngyTime.csv")
Hepta=pd.read_csv("Data/Hepta.csv")
Lsun=pd.read_csv("Data/Lsun.csv")
Target=pd.read_csv("Data/Target.csv")
Tetra=pd.read_csv("Data/Tetra.csv")
TwoDiamonds=pd.read_csv("Data/TwoDiamonds.csv")
WingNut=pd.read_csv("Data/WingNut.csv")
```

* Hepta (n = 212,var=3 classes = 7, clearly defined clusters)
* Tetra (n = 400,var=3, classes = 4, almost touching clusters)
* Chainlink (n = 1000,var=3 classes = 2, linearly non-separable)
* Atom (n = 800, var=3, classes = 2, different variances and linearly nonseparable)
* Lsun (n= 400, var=2 ,classes=3 ,different variances in clusters)
* EngyTime(n= 4096,var=2,classes= 2 ,density defined clusters)
* Target (n=770, var=2, classes=6, outliers)
* TwoDiamonds (n=800,var=2,classes= 2, touching clusters)
* WingNut(n=1070,var=2, classes=2, largest densities at cluster borders)

## Autoencoder 

Atom data

```python
print(df_atom.shape)
n = df_atom.shape[0]
p = df_atom.shape[1]
```

```python
inputL = Input(shape=(p,))
hiddenL = Dense(2,activation='sigmoid')(inputL)
outputL = Dense(p,activation='linear')(hiddenL)
autoencoder = Model(inputL,outputL)
```

La variable "autoencoder" repr√©sente le r√©seau dans son ensemble (le mod√®le), commen√ßant
par **inputL** et finissant par **outputL**. Nous en d√©finissons les caract√©ristiques d‚Äôapprentissage
en sp√©cifiant l‚Äôalgorithme d‚Äôoptimisation (optimizer = ‚Äòadam‚Äô) et le crit√®re √† optimiser (loss =
‚Äòmse‚Äô ; mean squared error).


```python
autoencoder.compile(optimizer='adam',loss='mse')
print(autoencoder.summary())
```

```python
n_epochs = 1000
historique = autoencoder.fit(x=df_atom,y=df_atom,epochs=n_epochs)
```

```python
plt.plot(np.arange(1,n_epochs+1),historique.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('D√©croissance fnct de perte')
plt.show()
```
